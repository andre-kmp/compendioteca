### 3.37

A função de jogo de computador collision() verifica se dois objetos circulares colidem; ela retorna True se elesmcolidirem e False se não colidirem. Cada objeto circular será dado pelo seu raio e as coordenadas (x, y) do seu centro.
Assim, a função apanhará seis números como entrada: as coordenadas (x1, y1) do centro e o raio r1 do primeiro círculo, e as coordenadas (x2, y2) do centro e o raio r2 do segundo círculo.

```python
>>> colisão(0, 0, 3, 0, 5, 3)
True
>>> colisão(0, 0, 1.4, 2, 2, 1.4)
False
```

##### Resolução

"Dois círculos estarão se sobrepondo se, e somente se, **a distância entre os dois centros** for menor do que **a soma dos dois raios**"

1. Calculando a distância entre os dois centros (Teorema de Pitágoras)

$d = \sqrt{(x_{2} - x_{1})^2+(y_{2}-y_{1})^2}$

2. Condições de sobreposição

- $d > r_{1} + r_{2}$ : não há contato
- $d = r_{1} + r_{2}$ : apenas se tocam
- $d < r_{1} + r_{2}$ : há sobreposição


````python
import matplotlib.pyplot as plt
import math

# Dados dos círculos - Inseridos pelo usuário

coordenadas_c1 = (0,0) # x,y
raio_c1 = 1.4

coordenadas_c2 = (2,2) # x,y
raio_c2 = 1.4

# Configuração das variáveis de trabalho

x1,y1,r1 = coordenadas_c1[0], coordenadas_c1[1], raio_c1
x2,y2,r2 = coordenadas_c2[0], coordenadas_c2[1], raio_c2

# TODO: Boa aplicação para um tradutor

def check_collision(x1,y1,r1,x2,y2,r2):
    d = math.dist((x1, y1),(x2, y2))
    sum_radius = r1+r2
    
    obj1 = plt.Circle((x1, y1), r1, color='g', alpha=0.6)
    obj2 = plt.Circle((x2,y2), r2, color='r', alpha=0.6)

    fig, ax = plt.subplots()

    ax.add_patch(obj1)
    ax.add_patch(obj2)

    plt.axis('square')
    plt.show()

    if d <= sum_radius:
        return True
    else: 
        return False

collision = check_collision(x1,y1,r1,x2,y2,r2)
print("Os círculos colidem") if collision else print("Os círculos não colidem")
````
